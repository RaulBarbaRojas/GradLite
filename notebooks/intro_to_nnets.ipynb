## 3. Artificial Intelligence, Machine learning and Deep Learning

### 3.1. Definitions
Artificial Intelligence (AI) is a **field of computer science focused on creating programs that can simulate human-like intelligence** (they can act, think, reason...).

Machine Learning (*ML*) is an AI paradigm that focuses on building algorithms that can **learn patterns from data** and **make predictions or decisions without being explicitly programmed** for every rule. Thus, *ML* develops **mathematical functions called models** (described in the next section) capable of **mapping specific inputs to outputs**. For example, a model might take an image as input and decide whether it depicts a cat or a dog.

Last, Deep Learning (*DL*) is a subset of machine learning that uses **artificial neural networks to automatically learn patterns from data and make predictions**. Thus, it leverages a **biological analogy** to achieve intelligent behavior.

### 3.2. Models and Types of Problems in ML/DL

A model in machine learning is essentially a **mathematical function that maps inputs to outputs**. These models have internal *parameters* (also called *weights*) that are involved in mathematical operations with the given inputs to generate the desired outputs. The **learning process adjusts the model's parameters**, so that the model produces the correct outputs for the introduced inputs. A model can be mathematically defined as follows:

$M_\theta: X \rightarrow Y, Y = M_\theta(X)$

Where:

- $M_\theta$ represents a model $M$ with parameters $\theta$. These are the parameters that are adjusted during the training of the model, so that the outputs are closer to their expected values.
- $X$ represents the model inputs. These could be the cat/dog images mentioned previously in the explanation, so that the model can use them as an input to decide whether the image depicts a cat or a dog.
- $Y$ represents the model output. In the previous example, it would imply be whether the image depicts a cat or a dog.

Depending on the task, **ML/DL models can solve different types of problems**:

- **Regression**: Predicting a continuous value from given inputs (e.g., predicting house prices based on size, location, inflation, etc.). To train a regression model, we need both the inputs and the expected outputs (e.g., real house prices in the previous example), so that the model can adjust its parameters accordingly to provide output values as close as possible to the expected values (e.g., the real house prices in the previous example).

- **Classification**: Assigning inputs to discrete categories.
Example: Determining whether an email is spam or not.

- Among other tasks, such as [clustering](https://en.wikipedia.org/wiki/Cluster_analysis), [anomaly detection](https://en.wikipedia.org/wiki/Anomaly_detection), etc.


Note that some of these tasks are called **supervised problems**, because we need input data and expected output data, but some other tasks are **non-supervised problems**. In the latter setting, we do not need (nor have) the expected outputs, that is, the model can only use the inputs to make its decisions (e.g., [clustering problems](https://en.wikipedia.org/wiki/Cluster_analysis)).

## 4. Neurons: The foundation of Deep Learning

### 4.1. Biological neurons

Deep learning is heavily inspired by the most complex and intelligente functioning unit that nature has ever produced: **biological neurons**.

**Biological neurons process and transmit information through electric and chemical signals**. The following figure provides a diagram of the structure of a biological neuron:

![A visual conceptualization of a biological neuron](images/bioneuron.jpg)

In the previous diagram:

- **Dendrites**: they are in charge of receiveing signals, so they are much like the inputs required to produce a response.
- **Synapses**: they are the junctions where one neuron connects to another, and they help regulate the strength (*weight*) of the input signals received from the dendrites.
- **Soma (cell body)**: this neuron component is responsible for integrating all the signals received from the dendrites. Thus, it has to **aggregate** the signals (*inputs*) received from the dendrites.
- **Axon**: the only responsibility of this component is to transmit the previous aggregation to other neurons, muscles or glands.


Thus, in summary, a biological neuron receives signals and process those signals in its body. If the signal is relevant, it will get *excited* and charge up electricity. If the electricity reaches a certain threshold, the neuron will allow this *excitement* to be transmitted to the next neuron. In the human brain, the *excitement threshold* varies from neuron to neuron, meaning that some neurons get *excited* about every signal, some other neurons hardly ever transmit *excitement*, showcasing an inhibitory effect.

### 4.2. Artificial Neurons

Deep learning's core concept is the **Artificial Neuron**, a **mathematical model that is heavily inspired by biological neurons**, capturing the core principles of signal reception, aggregation and propagation (much like the biological neurons!). The following is a diagram showcasing the structure of an artificial neuron:

![A visual conceptualization of an artificial neuron](images/artificial_neuron.webp)

In this conceptualization of an artificial neuron:

- **Inputs**: As can be seen, artificial neurons receive inputs, which we can consider *scalar values* (e.g., $x_1=0.5$). In this mathematical model, the inputs are analogous to the signals received by a biological neuron.
- **Weights**: also called *parameters*, they represent the relevance of each input for the specific artificial neuron. They are analogous to the synapses of biological neurons, determining the relevance of each of the input signals received by biological neurons. In this mathematical model that represents artificial neurons, the **input values** (e.g., $x_1$) **are multiplied by the corresponding weight** (e.g., $x_1 · w_1$, $x_2 · w_2$, ...), producing a weighted signal that will be processed by the *body* of the articial neuron.
- **Transfer function**: The transfer function represents the *body* of an artificial neuron. It is responsible for aggregating the weighted signals received by the neuron into a single value. This transfer function is, as a result, an **aggregation function of the different weighted signals**. While many aggregation functions exist (sum of all values, average of all values, etc.), the sum aggregation function is usually used as a transfer function. Furthermore, not only the weighted signals are summed, but the transfer function usually contains a bias parameter (**$b$**), which helps the artificial neuron have a higher or a lower *excitement threshold*. Thus, the bias helps some neuron be more *quick trigger* than others, just like it happens with biological neurons. The transfer function formula is expressed as follows: $\sum_{i=1}^{n}(w_i · x_i) + b$
- **Activation function**: last, the activation function of an artificial neuron is a mathematical function that determines whether the artificial neuron will *get excited*, analogous to the *excitement threshold* of biological neurons. In practical terms, many activation functions can be used (such as [ReLU](https://en.wikipedia.org/wiki/Rectified_linear_unit) and [sigmoid](https://en.wikipedia.org/wiki/Sigmoid_function)), but they all serve the same purpose: **introducing non-linearity**. In simple terms, **non-linearity means that the neuron's output is not just a straight-line combination of its inputs**. Without non-linearity, no matter how many layers a network has, the overall computation would always be **equivalent to a single linear transformation,** which severely **limits what the network can learn**. By applying a non-linear activation function, the neuron can model complex patterns, enabling deep neural networks to **solve complicated tasks like image recognition or language understanding**.
